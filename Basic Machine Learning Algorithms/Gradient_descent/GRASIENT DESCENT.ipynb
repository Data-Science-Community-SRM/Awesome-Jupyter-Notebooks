```python
import numpy as np
def gradient_descentwo(x,y):
    m_curr = b_curr = 0
    iterations = 1000
    lrate = 0.001
    n = len(x)
    for i in range(iterations):
        y_predi = m_curr * x + b_curr
        cost = (1/n) * sum([val**2 for val in (y-y_predi)])
        dm  = -(2/n) * sum(x * (y-y_predi))
        db  = -(2/n) * sum(y-y_predi)
        m_curr = m_curr - lrate * dm
        b_curr = b_curr - lrate * db 
        print("m{},b{},cost{},iteration{}".format(m_curr,b_curr,cost,i))
```

```python
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])
gradient_descentwo(x,y)
```
